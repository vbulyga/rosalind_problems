```python
def hamming_dist(seq_1, seq_2):
    dist = 0
    for i in range(len(seq_1)):
        dist += seq_1[i] != seq_2[i]
    return dist
```

```python
def rna_to_protein(rna_seq):
    idx_limit = len(rna_seq) - 2
    protein = ''
    codon = '---'
    idx = 0
    while(idx < idx_limit):
        try:
            codon = rna_seq[idx:idx + 3]
            # print(codon, idx)
            idx += 3
        except:
            print('error')
            break
        match codon: 
            case 'GCU' | 'GCC' | 'GCA' | 'GCG':
                protein += 'A'
            case 'UGU' | 'UGC':
                protein += 'C'
            case 'GAU' | 'GAC':
                protein += 'D'
            case 'GAA' | 'GAG':
                protein += 'E'
            case 'UUU' | 'UUC':
                protein += 'F'
            case 'GGU' | 'GGC' | 'GGA' | 'GGG':
                protein += 'G'
            case 'CAU' | 'CAC':
                protein += 'H'
            case 'AUU' | 'AUC' | 'AUA':
                protein += 'I'
            case 'AAA' | 'AAG':
                protein += 'K'
            case 'UUA' | 'UUG' | 'CUU' | 'CUC' | 'CUA' |'CUG':
                protein += 'L'
            case 'AUG':
                protein += 'M'
            case 'AAU' | 'AAC':
                protein += 'N'
            case 'CCU' | 'CCC' | 'CCA' | 'CCG':
                protein += 'P'
            case 'CAA' | 'CAG':
                protein += 'Q'
            case 'CGU' | 'CGC' | 'CGA' | 'CGG' | 'AGA' | 'AGG':
                protein += 'R'
            case 'UCU' | 'UCC' | 'UCA' | 'UCG' | 'AGU' | 'AGC':
                protein += 'S'
            case 'ACU' | 'ACC' | 'ACA' | 'ACG':
                protein += 'T'
            case 'GUU' | 'GUC' | 'GUA' | 'GUG':
                protein += 'V'
            case 'UGG':
                protein += 'W'
            case 'UAU' | 'UAC':
                protein += 'Y'
            case 'UAA' | 'UAG' | 'UGA':
                print(idx)
                break
    return protein
```
